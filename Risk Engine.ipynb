# 1.Supervised Learning

###### Train-Test Split 

```python
#X=df.drop(["Genuine"],axis=1)
#Y=df["Genuine"]

X=df.drop(['Login_Time', 'IP address', 'City', 'Region', 'Browser', 'Version',

'OS', 'OS_Version', 'DI2', 'DI3', 'Provider', 'H.period','H.e', 'DD.e.five', 'UD.e.five','DD.five.Shift.r',

'UD.five.Shift.r','Genuine'],axis=1)
Y=df["Genuine"]
```

```python
from sklearn.model_selection import train_test_split 
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X, Y, test_size = 0.3)
```

## SVM

```python
#SVM
# import SVM libraries 
from sklearn.svm import SVC
from sklearn.metrics import classification_report, roc_auc_score

model=SVC()
clf= model.fit(X_Train, Y_Train)
pred_SVM = clf.predict(X_Test)
```

######  Model performance on test set

```python
from sklearn.metrics import classification_report
print(classification_report(Y_Test, pred_SVM))
```

###### K-Fold Cross validation to obtain Average Test accuracy and standard deviation

```python

from numpy import mean
from numpy import std
from sklearn.datasets import make_classification
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score



cv = KFold(n_splits=10, random_state=1, shuffle=True)
# create model

# evaluate model
scores = cross_val_score(clf, X_Test, Y_Test, scoring='accuracy', cv=cv, n_jobs=-1)
# report performance
print('Accuracy: %.3f (%.3f)' % (mean(scores), std(scores)))
```
